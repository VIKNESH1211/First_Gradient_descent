```python
import numpy as np
```

```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
import matplotlib.pyplot as plt
plt.scatter(x,y,c='r')
```

```python
print(x)
print("_________________________________________________")
print(y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
ols_pred = model.predict(x)

ols_pred
```

```python
model.intercept_
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,c='r',label = 'Actual values')
plt.plot(x,ols_pred,c='b',label='best fit line')
plt.legend()
```

```python
# now using gradient descent algo
# gradient decent is a optimiser
# the basic of adam is gradient desent 
```

```python
# Loss L = c^2 - relationship which gives the gradient descent curve
# we have to find c whre L is minimum
```

```python
# equation for gradient descent - c'new = c'old - slope
```

```python
#step 1 : select a random value of c , at this point loss is high
```

```python
# when c is initialised as a negative number

# if c value is decreasing , we are going far from the minimum
# if c value is increasing , we are going close from the minimum

# when c is initialised as a positive number

# if c value is decreasing , we are going close from the minimum
# if c value is increasing , we are going for from the minimum
```

```python
# keep the slope value constant at m = 78.35063668
# c = 26.15963284313262 is the optimal value (we need to read this c value using the is)
```

```python
#let c = 0
y_pred = ((78.35*x)+0)
y_pred 
# reshape these values to plot in single line
y_pred = ((78.35*x)+0).reshape(4)
y_pred
```

```python
plt.scatter(x,y,c='r',label = 'Actual values')
plt.plot(x,ols_pred,c='black',label='ols')
plt.plot(x,y_pred,c='blue',label='c=0')
plt.legend() 
```

```python
m = 78.35
c = 0 
loss_slope = -2 * np.sum( y - m * x.ravel() - c) #to calculate slope at c =0
loss_slope
```

```python
# learning rate (consider lr = 0.1)
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# calculating new c value
c = c - step_size
c #c value jumped from 0 to 20.92
```

```python
# plot the line for the updated c value
y_pred1 = ((78.35 * x) + c).reshape(4)
plt.figure(figsize =(10,10))
plt.scatter(x,y,c='r',label = 'Actual values')
plt.plot(x,ols_pred,c='black',label='ols')
plt.plot(x,y_pred,c='blue',label='c=0')
plt.plot(x,y_pred1,c='red',label='c=20.92')
plt.legend() 
```

```python
loss_slope = -2 * np.sum( y - m * x.ravel() - c) #to calculate slope 
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
#updated c value(intercept)
c = c - step_size
c
```

```python
#plot the line for the updated C value
y_pred2 = ((78.35 * x) + c).reshape(4)
plt.figure(figsize =(10,10))
plt.scatter(x,y,c='r',label = 'Actual values')
plt.plot(x,ols_pred,c='black',label='ols')
plt.plot(x,y_pred,c='blue',label='c=0')
plt.plot(x,y_pred1,c='red',label='c=20.92')
plt.plot(x,y_pred2,c='pink',label='c=25.11')
plt.legend() 
```

```python
# using a loop for gradient descent
c = - 100
m = 78.35
lr = 0.1

epochs = 10
plt.figure(figsize=(10,10))

for i in range(epochs):
  loss_slope = -2 * np.sum( y - m * x.ravel() - c) # to calculate the slope
  c =  c- (lr*loss_slope)
  y_pred = m * x + c
  plt.plot(x,y_pred)

plt.scatter(x,y)
```
